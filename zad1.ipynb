```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python

def calculate_omegas(M1, M2, Y1, Y2, a, q):
    alpha1 = ((Y1 + Y2)*(M1 + M2) + np.sqrt((Y1 + Y2)**2 * (M1 + M2)**2 - 4*M1*M2*(2*Y1*Y2)*(1 - np.cos(q*a))))/(2*M1*M2)
    alpha1 = np.abs(alpha1)
    
    alpha2 = ((Y1 + Y2)*(M1 + M2) - np.sqrt((Y1 + Y2)**2 * (M1 + M2)**2 - 4*M1*M2*(2*Y1*Y2)*(1 - np.cos(q*a))))/(2*M1*M2)
    alpha2 = np.abs(alpha2)

    return np.sqrt(alpha1), np.sqrt(alpha2)
```

```python


#constants

_a = 1

# _M1 = 4
# _M2 = 1
# _Y1 = 2
# _Y2 = 3

# _M1 = 3
# _M2 = 2
# _Y1 = 2
# _Y2 = 1

# _M1 = 2
# _M2 = 3
# _Y1 = 3
# _Y2 = 4

_M1 = 1
_M2 = 4
_Y1 = 4
_Y2 = 2
 
# _M1 = 4
# _M2 = 4
# _Y1 = 1
# _Y2 = 5

df = pd.DataFrame({
    'q' : np.arange(-np.pi/_a, np.pi/_a, 0.01),
})

df['omega1'], df['omega2'] = calculate_omegas(_M1, _M2, _Y1, _Y2, _a, df['q'])

df
```

```python
fig, ax = plt.subplots(1, figsize = (8,6), dpi=150)

_linewidth = 1
_color = 'green'
ax.plot(df['q'], df['omega1'], linewidth = _linewidth, color = _color)

_linewidth = 1
_color = 'pink'
ax.plot(df['q'], df['omega2'], linewidth = _linewidth, color = _color)

plt.xlabel(r'$q \; [\frac{1}{a}]$')
plt.ylabel(r'$\omega \; [a.u.]$')

plt.savefig('out/zad1_wyk_omega_od_q.jpg', dpi=350)
plt.show()
```
